**MEMERIKSA DATA**

```python
import pandas as pd

# Memuat data dari file CSV
df = pd.read_csv('movie_sample_dataset.csv')

# Menampilkan 5 baris pertama dari dataset
df. head()
```

```python
import numpy as np

# Mengganti "?" dengan NaN
df.replace("?", np.nan, inplace=True)

# Memeriksa jumlah missing values di setiap kolom
print(df.isnull().sum())

# Memeriksa tipe data
df.dtypes
```

**MEMBERSIHKAN DATA**

HAPUS BARIS YANG MEMUAT NILAI NaN ATAU KOSONG

```python

# Menghapus baris dengan missing values pada kolom

df.dropna(subset=["gross", "budget"], axis=0, inplace=True)

# Reset index setelah menghapus baris
df.reset_index(drop=True, inplace=True)
```

```python
# Menghapus baris dengan missing values pada kolom

df.dropna(subset=["color", "director_name", "genres"], axis=0, inplace=True)

# Reset index setelah menghapus baris
df.reset_index(drop=True, inplace=True)
```

MENCARI DAN MENGUBAH NILAI NEGATIF

```python
import pandas as pd
#mencari nilai negatif
for col in df.select_dtypes(include=["int", "float"]).columns:
    print(f"\nKolom {col}:")
    print(df[df[col] < 0])
```

```python
#menghilangkan nilai negatif
df['duration'] = df['duration'].apply(lambda x: abs(x) if x < 0 else x)
df['imdb_score'] = df['imdb_score'].apply(lambda x: abs(x) if x < 0 else x)
```

TRANSFORMASI DATA

UBAH TIPE DATA

```python

# Ubah kolom 'gross' dan 'budget' menjadi numerik
df['gross'] = pd.to_numeric(df['gross'])
df['budget'] = pd.to_numeric(df['budget'])
```

SPLIT DAN EXPLODE GENRES

```python
df['genres'] = df['genres'].str.split('|')
df = df.explode('genres')
df.head()
```

UBAH SEMUA TEXT JADI UPPERCASE

```python
# Mengubah semua kolom teks menjadi huruf besar
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
```

SIMPAN FILE CLEANED

```python
df.to_csv('movie_dataset_cleaned.csv', index=False)
```
